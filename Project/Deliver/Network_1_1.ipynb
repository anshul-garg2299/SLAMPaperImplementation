```python
import tensorflow as tf
import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,LSTM
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#from sklearn.preprocessing import MinMaxScalar,LabelEncoder
```

```python
data = pd.read_csv("task1_noise.csv", header=None, names=['x0','y0','x1','y1','x2','y2','x3','y3','x4','y4','x5','y5','x6','y6','x7','y7','x8','y8','x9','y9','speed','vangle','index','d','beta','true_x','true_y'])
samples=[]
length = 5000
step = 6
for i in range(0,length*step,step):
    sample=np.array(data[i:(i+step)])
    samples.append(sample)
data=np.asarray(samples)
np.random.shuffle(data)

truth = []
for i in range(length):
    truth.append([data[i][0][25],data[i][0][26]])
    
truth = np.asarray(truth)
data = data[:,:,:25]

truth_train = truth[:4000,:]
data_train = data[:4000,:,:]
truth_test = truth[4000:,:]
data_test = data[4000:,:,:]
```

```python
model = Sequential()
model.add(LSTM(256,input_shape=(6,25)))
model.add(Dense(2))
#model.summary()
model.compile(loss='mse',metrics=['accuracy'],optimizer = 'adam')
model.fit(data_train,truth_train,epochs =3, batch_size = 1,validation_data = (data_test,truth_test))

score = model.evaluate(data_test,truth_test)
print(score)
```
